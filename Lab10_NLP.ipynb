```python
def bow(texts):
    unique_words = set()
    vocabulary_ = {}

    # Подсчет уникальных слов в текстах и создание словаря
    for text in texts:
        unique_words.update(text.split())
        
    # Пронумерование слов в словаре
    for i, word in enumerate(unique_words):
        vocabulary_[word] = i

    # Создание матрицы Bag of Words
    bow_matrix = np.zeros((len(texts), len(unique_words)), dtype=int)

    # Заполнение матрицы Bag of Words
    for i, text in enumerate(texts):
        words = text.split()
        for word in words:
            col_idx = vocabulary_[word]
            bow_matrix[i, col_idx] += 1

    return unique_words, bow_matrix, vocabulary_
```

```python
def tf_idf(texts):
    word_count_per_doc = {}
    doc_count_per_word = {}
    unique_words = set()

    # Подсчет частоты слов и формирование уникального словаря
    for i, text in enumerate(texts):
        word_count_per_doc[i] = {}
        words = text.split()
        unique_words.update(words)
        for word in words:
            if word_count_per_doc[i].get(word):
                word_count_per_doc[i][word] += 1
            else:
                word_count_per_doc[i][word] = 1
            if word in doc_count_per_word:
                doc_count_per_word[word].add(i)
            else:
                doc_count_per_word[word] = {i}

    # Создание массива для хранения векторов TF-IDF
    tf_idf_matrix = np.zeros((len(texts), len(unique_words)))

    # Вычисление TF-IDF
    for i, text in enumerate(texts):
        words = text.split()
        total_words = len(words)
        for word in set(words):
            word_count = word_count_per_doc[i][word]
            word_index = list(unique_words).index(word)
            tf = word_count / total_words
            idf = math.log((len(texts) + 1) / (len(doc_count_per_word[word]) + 1) + 1)
            tf_idf_matrix[i, word_index] = tf * idf

    return list(unique_words), tf_idf_matrix
```

```python
def my_pca(X, n_components=2):
    mean = np.mean(X, axis=0)
    X_centered = X - mean
    cov_matrix = np.cov(X_centered.T)
    values, vectors = np.linalg.eig(cov_matrix)
    sorted_indices = np.argsort(values)[::-1]
    sorted_values = values[sorted_indices]
    sorted_vectors = vectors[:, sorted_indices]
    selected_vectors = sorted_vectors[:, :n_components]
    X_projected = X_centered.dot(selected_vectors)
    return X_projected
```

```python
X_pca = my_pca(X, n_components=5)
```

```python
class Kmeans:
    def __init__(self, n_clusters, max_iters=100):
        self.n_clusters = n_clusters
        self.max_iters = max_iters

    def initialize_centroids(self, X):
        centroids_idx = np.random.choice(range(X.shape[0]), size=self.n_clusters, replace=False)
        centroids = X[centroids_idx]
        return centroids

    def assign_clusters(self, X, centroids):
        distance_matrix = np.zeros((X.shape[0], self.n_clusters))
        for i, centroid in enumerate(centroids):
            distance_matrix[:, i] = np.linalg.norm(X - centroid, axis=1)
        cluster_labels = np.argmin(distance_matrix, axis=1)
        return cluster_labels

    def update_centroids(self, X, cluster_labels):
        centroids = np.zeros((self.n_clusters, X.shape[1]))
        for i in range(self.n_clusters):
            centroids[i] = np.mean(X[cluster_labels == i], axis=0)
        return centroids

    def fit(self, X):
        centroids = self.initialize_centroids(X)
        for _ in range(self.max_iters):
            cluster_labels = self.assign_clusters(X, centroids)
            new_centroids = self.update_centroids(X, cluster_labels)
            if np.allclose(new_centroids, centroids):
                break
            centroids = new_centroids
        return centroids, cluster_labels
```

```python
from time import time
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                max_features=1000,
                                stop_words='english')
tf = tf_vectorizer.fit_transform(df['v3'])
```

```python
lda = LatentDirichletAllocation(n_components=25, max_iter=20,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0).fit(tf)
```

```python
tf_vectorizer.get_feature_names_out()
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
```

```python
print_top_words(lda, tf_vectorizer.get_feature_names_out(), 10)
```
