Обработаем выбросы

Создадим датафрейм со столбцами с выбросами 

```python
emissions = df[['audience_count', 'tomatometer_rotten_critics_count', 'streaming_release_year']]
emissions
```

Определим нижние приделы

```python
Q1 = emissions.quantile(0.25)
Q1
```

Определим верхние приделы

```python
Q3 = emissions.quantile(0.75)
Q3
```

Составимм IQR оценки

```python
IQR = Q3-Q1
```

Отфильтруем датафрейм от выбросов 

```python
df_filtered = emissions[~((emissions < (Q1 - 1.5 * IQR)) |(emissions > (Q3 + 1.5 * IQR))).any(axis=1)]
df_filtered
```

Скопируем индексы

```python
index_list = list(df_filtered.index.values)
```

И теперь данные без выбросов сохраним в отдельном датафрейме 

```python
df_filtered = df[df.index.isin(index_list)]
df_filtered.head()
```

теперь создадим гистограммы без выбросов

```python
round(corr_MGM,3)
```
